```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv(r"2019.csv")
hired = data[data["hired"]]
easy = hired[hired['easy_apply']]
normal = hired[~hired['easy_apply']]


numEasy = len(easy)
numNormal = len(normal)
totals = [numEasy, numNormal]
labels=["Hired by Easy Apply", "Hired without Easy Apply"]
colors=["#F99DF9", "#7AF711"]

plt.pie(totals, labels=labels, autopct='%1.1f%%', shadow=True, startangle=1200, colors=colors)
plt.title("Hired with/without Easy Apply")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv(r"2019.csv")
hired = data[data["hired"]]
easy = hired[hired['easy_apply']]
normal = hired[~hired['easy_apply']]

notHired = data[~data["hired"]]
nhEasy = notHired[notHired['easy_apply']]
nhNormal = notHired[~notHired['easy_apply']]

numEasy = len(easy)
numNormal = len(normal)
numNhEasy = len(nhEasy)
numNhNormal = len(nhNormal)

totals = [numEasy, numNhEasy, numNormal, numNhNormal]
labels=["Hired, Used Easy Apply", "Not Hired, Used Easy Apply", "Hired, Did Not Use Easy Apply", "Not Hired, Did Not Use Easy Apply",]
colors=["#FECDFF", "#95D8FF", "#A1F69F", "#FBFE7B"]

plt.pie(totals, labels=labels, autopct='%1.1f%%', shadow=True, startangle=150, colors=colors)
plt.title("Hired/Not Hired with/without Easy Apply")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded the data from the CSV file
data = pd.read_csv(r"2019.csv")

# Assuming "hired" is a column representing whether an individual was hired
hired = data[data["hired"]]

# Assuming "job_location" is the column containing job locations
hired['job_location'].value_counts().plot(kind='bar', figsize=(12, 6))
plt.title('Distribution of Job Locations for Hired Individuals')
plt.xlabel('Job Location')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded the data from the CSV file
data = pd.read_csv("2019.csv")

# Assuming "hired" is a column representing whether an individual was hired
hired = data[data["hired"]]

# Assuming "job_location" is the column containing job locations
locations = ['New York, NY', 'Bronx, NY', 'Brooklyn, NY']
hired_locations = [hired[hired['job_location'] == location] for location in locations]

location_counts = pd.concat([hired_location['job_location'].value_counts() for hired_location in hired_locations], axis=1)
location_counts.columns = locations

location_counts.plot(kind='bar',width=2)
plt.title('Distribution of Hired Individuals in NY')
plt.xlabel('Job Location')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("2019.csv")


hired = data[data["hired"]]


locations = ['New York, NY', 'Bronx, NY', 'Brooklyn, NY']
hired_locations = [hired[hired['job_location'] == location] for location in locations]

location_counts = pd.concat([hired_location['job_location'].value_counts() for hired_location in hired_locations], axis=1)
location_counts.columns = locations

location_counts.plot(kind='bar', width=2)
plt.title('Distribution of Hired Individuals in NY')
plt.xlabel('Job Location')
plt.ylabel('Count')


for index, value in enumerate(location_counts.values):
    for i, v in enumerate(value):
        if not pd.isna(v):  # Check if the value is not NaN
            plt.text(index + i * 0.2 - 0.2, v + 0.2, str(v), color='black', fontweight='bold')

plt.show()
```
